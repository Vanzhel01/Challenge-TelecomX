```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#ðŸ“Œ ExtracciÃ³n

```python
clientes=pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
clientes=pd.DataFrame(clientes)
clientes
```

```python
pd.json_normalize(clientes['customer'])
clientes=pd.concat([clientes,pd.json_normalize(clientes['customer'])],axis=1)
clientes.drop('customer',axis=1,inplace=True)
clientes
```

```python
clientes=pd.concat([clientes,pd.json_normalize(clientes['phone'])],axis=1)
clientes.drop('phone',axis=1,inplace=True)
clientes
```

```python
clientes=pd.concat([clientes,pd.json_normalize(clientes['internet'])],axis=1)
clientes.drop('internet',axis=1,inplace=True)
clientes
```

```python
clientes=pd.concat([clientes,pd.json_normalize(clientes['account'])],axis=1)
clientes.drop('account',axis=1,inplace=True)
clientes
```

```python
clientes['Churn'] = clientes['Churn'].map({'No':0,'Yes':1,'':0}).fillna(0).astype(int)
clientes
```

```python
pd.unique(clientes['Churn'])
```

```python
clientes['Cuentas Diarias']=clientes['Charges.Monthly']/30
clientes
```

```python
clientes['Partner'] = clientes['Partner'].map({'No':0,'Yes':1}).fillna(0).astype(int)
clientes['Dependents'] = clientes['Dependents'].map({'No':0,'Yes':1}).fillna(0).astype(int)
clientes['PhoneService'] = clientes['PhoneService'].map({'No':0,'Yes':1}).fillna(0).astype(int)
clientes['MultipleLines'] = clientes['MultipleLines'].map({'No':0,'Yes':1,'No phone service':0}).fillna(0).astype(int)
clientes['OnlineSecurity'] = clientes['OnlineSecurity'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['OnlineBackup'] = clientes['OnlineBackup'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['DeviceProtection'] = clientes['DeviceProtection'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['TechSupport'] = clientes['TechSupport'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['StreamingTV'] = clientes['StreamingTV'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['StreamingMovies'] = clientes['StreamingMovies'].map({'No':0,'Yes':1,'No internet service':0}).fillna(0).astype(int)
clientes['PaperlessBilling'] = clientes['PaperlessBilling'].map({'No':0,'Yes':1}).fillna(0).astype(int)
```

```python
clientes.rename(columns={'SeniorCitizen':'Mayor de 65 aÃ±os','Partner':'Pareja','Dependents':'Dependientes','tenure':'Meses de antigÃ¼edad','PaperlessBilling':'FacturaciÃ³n Electronica','Charges.Monthly':'Tarifa Mensual','Charges.Total':'Total Pagado'},inplace=True)
clientes
```

### AnÃ¡lisis de las distribuciones de 'Meses de antigÃ¼edad' y 'Total Pagado' en relaciÃ³n con el Churn

**1. DistribuciÃ³n de 'Meses de antigÃ¼edad' por Churn:**

**ConclusiÃ³n sobre 'Meses de antigÃ¼edad':** Los clientes que permanecen poco tiempo son los mÃ¡s propensos a *churnear*, mientras que aquellos con una lealtad mÃ¡s larga tienden a quedarse. Es crucial enfocarse en la retenciÃ³n de nuevos clientes.

**2. DistribuciÃ³n de 'Total Pagado' por Churn:**
*   **Clientes sin Churn (0):** La distribuciÃ³n muestra una densidad mÃ¡s alta para clientes con pagos totales tanto bajos como muy altos, con una tendencia general a tener pagos totales mÃ¡s elevados. Esto podrÃ­a indicar que tanto los clientes de bajo costo que no experimentan problemas como los clientes premium estÃ¡n menos inclinados al churn.
*   **Clientes con Churn (1):** La distribuciÃ³n para clientes con churn es notablemente mÃ¡s alta en el rango de pagos totales bajos a medios. **Consideraciones Generales:**

### Variables CategÃ³ricas Identificadas para el AnÃ¡lisis de Churn

Las siguientes columnas han sido identificadas como variables categÃ³ricas (incluyendo las binarias 0/1) relevantes para el anÃ¡lisis de churn en el DataFrame `clientes`:

*   `gender`
*   `Mayor de 65 aÃ±os`
*   `Pareja`
*   `Dependientes`
*   `PhoneService`
*   `MultipleLines`
*   `InternetService`
*   `OnlineSecurity`
*   `OnlineBackup`
*   `DeviceProtection`
*   `TechSupport`
*   `StreamingTV`

*   `FacturaciÃ³n Electronica`

```python
categorical_cols = ['gender', 'Mayor de 65 aÃ±os', 'Pareja', 'Dependientes', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'FacturaciÃ³n Electronica', 'PaymentMethod']
```

```python
for col in categorical_cols:
    churn_rate = clientes.groupby(col)['Churn'].value_counts(normalize=True).unstack()
    if 1 in churn_rate.columns:
        churn_rate = churn_rate[1] * 100
    else:
        churn_rate = pd.Series(0, index=churn_rate.index)

    plt.figure(figsize=(8, 5))
    sns.barplot(x=churn_rate.index, y=churn_rate.values, hue=churn_rate.index, palette='viridis', legend=False)


    plt.title(f'Tasa de Churn por {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Tasa de Churn (%)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## Limpieza y Tratamiento de Datos

### Proceso de Limpieza y PreparaciÃ³n de Datos

Para preparar el conjunto de datos para el anÃ¡lisis, se llevaron a cabo los siguientes pasos de limpieza y tratamiento:

1.  **ImportaciÃ³n de Datos**: Los datos se cargaron inicialmente en un DataFrame de pandas utilizando `pd.read_json()` directamente desde la URL proporcionada, lo que permitiÃ³ un acceso rÃ¡pido y eficiente a la informaciÃ³n.

2.  **Desanidamiento de Columnas JSON**: El DataFrame original contenÃ­a columnas con estructuras JSON anidadas ('customer', 'phone', 'internet', 'account'). Para facilitar el anÃ¡lisis, estas columnas se desanidaron utilizando `pd.json_normalize()`. Las nuevas columnas resultantes de este proceso se concatenaron con el DataFrame principal, y las columnas JSON originales se eliminaron para evitar redundancia y mejorar la estructura del DataFrame.

3.  **Manejo de 'Charges.Total'**: La columna 'Charges.Total' fue identificada como un tipo de objeto (`object`) que contenÃ­a cadenas vacÃ­as (`''`). Para convertirla a un tipo numÃ©rico, se realizÃ³ lo siguiente:
    *   Las cadenas vacÃ­as fueron reemplazadas por valores nulos (`np.nan`) utilizando `replace(r'^\s*$', np.nan, regex=True)`.
    *   Posteriormente, la columna se convirtiÃ³ a tipo numÃ©rico (`float`) utilizando `pd.to_numeric(errors='coerce')`, manejando cualquier error de conversiÃ³n al establecerlos tambiÃ©n como `NaN`.
    *   Finalmente, los valores `NaN` resultantes se rellenaron con `0` utilizando `.fillna(0)`, lo que asegurÃ³ que todos los valores en la columna fueran numÃ©ricos y manejables.

4.  **ConversiÃ³n de Columnas NumÃ©ricas**: AdemÃ¡s de 'Charges.Total', columnas como 'Meses de antigÃ¼edad' (`tenure`) y 'Tarifa Mensual' (`Charges.Monthly`) fueron confirmadas como tipos numÃ©ricos (`int64` y `float64` respectivamente), por lo que no requirieron conversiones adicionales en esta etapa.

5.  **CodificaciÃ³n Binaria de Variables CategÃ³ricas**: Varias columnas categÃ³ricas que representaban opciones binarias (SÃ­/No) o categorÃ­as que podÃ­an interpretarse como binarias se convirtieron a una representaciÃ³n numÃ©rica (0/1) para facilitar el anÃ¡lisis y el modelado. Esto incluyÃ³:
    *   `Churn`: Convertida de {'No':0, 'Yes':1, '':0}.
    *   `Partner`, `Dependents`, `PhoneService`, `PaperlessBilling`: Mapeadas de {'No':0, 'Yes':1}.
    *   `MultipleLines`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV`, `StreamingMovies`: Mapeadas de {'No':0, 'Yes':1, 'No phone service':0} o {'No internet service':0} segÃºn corresponda, para unificar las representaciones binarias.
